### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(6, 6))  # Adjust figure size if needed

plt.scatter(x, y)

for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Scatter Plot with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True) #Optional: Adds a grid to the plot
plt.show()
```
