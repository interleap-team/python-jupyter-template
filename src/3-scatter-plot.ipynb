### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='blue')

# Annotate points
for i, txt in enumerate(zip(x, y)):
    plt.annotate(f"{txt}", (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Add title and labels
plt.title("Scatter Plot with Annotations")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

# Show plot
plt.show()
```
