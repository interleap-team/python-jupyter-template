### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt
import numpy as np


# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, color='orange', label='Data Points')

# Annotate each point
for i, (x_val, y_val) in enumerate(zip(x_scatter, y_scatter)):
    plt.annotate(f'({x_val}, {y_val})', (x_val, y_val), textcoords="offset points", xytext=(5, 5), fontsize=10)

# Customize the scatter plot
plt.title('Scatter Plot with Annotations', fontsize=16)
plt.xlabel('x values', fontsize=14)
plt.ylabel('y values', fontsize=14)
plt.legend(fontsize=12)

# Show the scatter plot
plt.show()
```
