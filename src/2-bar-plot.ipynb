### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.

import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Compute y values as square of x

# Create the line plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x^2', color='b', linewidth=2)

# Customize the line plot
plt.title("Line Plot of y = x^2")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the line plot
plt.show()

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['r', 'g', 'b', 'y']

# Create the bar plot
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)

# Customize the bar plot
plt.title("Bar Plot of Categories")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the bar plot
plt.show()
```
