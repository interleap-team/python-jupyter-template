### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green' , 'purple']
plt.bar(categories, values, color=colors)
plt.title("Bar Plot witj Different Colors")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
