### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error: {e}. Install matplotlib.")
    exit()

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

if len(categories) != len(values):
    print("Error: Mismatch between categories and values.")
    exit()

plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'])
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

```
