### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
try:
    import matplotlib.pyplot as plt
    import numpy as np
except ImportError as e:
    print(f"Error: {e}. Install matplotlib and numpy.")
    exit()

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line Plot
x = np.linspace(0, 10, 100)
y = x**2
if x is None or y is None:
    print("Error: Missing data for line plot.")
    exit()
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Plot')

# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
if len(categories) != len(values):
    print("Error: Mismatch between categories and values.")
    exit()
axs[0, 1].bar(categories, values, color='green')
axs[0, 1].set_title('Bar Plot')

# Scatter Plot
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
if len(scatter_x) != len(scatter_y):
    print("Error: Scatter plot data mismatch.")
    exit()
axs[1, 0].scatter(scatter_x, scatter_y, color='red')
axs[1, 0].set_title('Scatter Plot')

# Pie Chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
if sum(sizes) == 0:
    print("Error: Pie chart sizes cannot be all zero.")
    exit()
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()


```
