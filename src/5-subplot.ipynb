### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.arange(1, 6)
y = x * 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot (Top-left)
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = 2x')
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('X Values')
axes[0, 0].set_ylabel('Y Values')
axes[0, 0].grid(True, linestyle='--', alpha=0.6)
axes[0, 0].legend()

# Bar plot (Top-right)
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Scatter plot (Bottom-left)
axes[1, 0].scatter(x, y, color='blue', marker='o', label='Data Points')
for i in range(len(x)):
    axes[1, 0].text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=10, ha='right', va='bottom')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X Values')
axes[1, 0].set_ylabel('Y Values')
axes[1, 0].grid(True, linestyle='--', alpha=0.6)
axes[1, 0].legend()

# Pie chart (Bottom-right)
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'purple'], startangle=140)
axes[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
