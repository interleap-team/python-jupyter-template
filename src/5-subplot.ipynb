### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
bar_labels = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['pink', 'grey', 'lightcoral', 'lightskyblue']
pie_explode = (0.1, 0, 0, 0)  # explode 1st slice

fig, axs = plt.subplots(2, 2, figsize=(10, 10))


axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart')


axs[0, 1].bar(bar_labels, bar_values, color=['pink', 'skyblue', 'grey', 'purple'])
axs[0, 1].set_title('Bar Chart')


axs[1, 0].scatter(scatter_x, scatter_y)
axs[1, 0].set_title('Scatter Plot')


axs[1, 1].pie(pie_sizes, explode=pie_explode, labels=pie_labels, colors=pie_colors, autopct='%1.1f%%', shadow=True, startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  

plt.tight_layout()
plt.show()
```
