### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = x * 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Chart (Top Left)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = 2x')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X Values")
axs[0, 0].set_ylabel("Y Values")
axs[0, 0].grid(True)
axs[0, 0].legend()

# Bar Chart (Top Right)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Scatter Plot (Bottom Left)
axs[1, 0].scatter(x, y, color='blue', marker='o')
for i in range(len(x)):
    axs[1, 0].annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X Values")
axs[1, 0].set_ylabel("Y Values")
axs[1, 0].grid(True)

# Pie Chart (Bottom Right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, shadow=True, colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
