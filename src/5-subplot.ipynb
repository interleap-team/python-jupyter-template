### Exercise 5: Subplot Layout

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [25, 15, 30, 30]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 1, 5, 3]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]


# Create the subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

# Plot in the first subplot (line chart)
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Line Chart')

# Plot in the second subplot (bar chart)
axes[0, 1].bar(categories, values)
axes[0, 1].set_title('Bar Chart')

# Plot in the third subplot (scatter plot)
axes[1, 0].scatter(x_scatter, y_scatter)
axes[1, 0].set_title('Scatter Plot')

# Plot in the fourth subplot (pie chart)
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 1].axis('equal')  # Equal aspect ratio for pie chart
axes[1, 1].set_title('Pie Chart')

# Add a general title to the entire figure (optional)
fig.suptitle('Subplots Example', fontsize=16)

# Adjust spacing between subplots for better layout (optional)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust top to make space for suptitle

# Display the plots
plt.show() 
```
