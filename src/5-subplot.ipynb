### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Data for the plots
x = [1, 2, 3, 4, 5]
y_line = [1, 3, 2, 5, 4]
y_bar = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line chart
axs[0, 0].plot(x, y_line, marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Second subplot: Bar chart
axs[0, 1].bar(['A', 'B', 'C', 'D'], y_bar, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Third subplot: Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter)
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0,5), ha='center')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie chart is a circle
axs[1, 1].set_title('Programming Language Popularity')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
