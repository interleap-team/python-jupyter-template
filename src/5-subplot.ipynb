### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

scatter_x = np.random.rand(20)
scatter_y = np.random.rand(20)

pie_sizes = [20, 30, 25, 25]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (Top-left)
axs[0, 0].plot(x, y, label='Sine Wave', color='b')
axs[0, 0].set_title("Line Chart")
axs[0, 0].legend()

# Bar chart (Top-right)
axs[0, 1].bar(categories, values, color='g')
axs[0, 1].set_title("Bar Chart")

# Scatter plot (Bottom-left)
axs[1, 0].scatter(scatter_x, scatter_y, color='r', marker='o')
axs[1, 0].set_title("Scatter Plot")

# Pie chart (Bottom-right)
axs[1, 1].pie(pie_sizes, labels=categories, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
