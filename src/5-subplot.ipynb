### Exercise 5: Subplot Layout

```python
# TODO: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot a line chart in the first subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title("Line Chart: y = x^2")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("y = x^2")

# Plot a bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Plot a scatter plot in the third subplot
axs[1, 0].scatter(x, y)
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("y")

# Plot a pie chart in the fourth subplot
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Pie Chart: Language Popularity")

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
