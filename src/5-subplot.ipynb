### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.arange(1, 6)
y = x * 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 1st subplot: Line chart
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X-axis")
axes[0, 0].set_ylabel("Y-axis")

# 2nd subplot: Bar chart
axes[0, 1].bar(categories, values, color=['r', 'g', 'b', 'y'])
axes[0, 1].set_title("Bar Chart")

# 3rd subplot: Scatter plot
axes[1, 0].scatter(x, y, color='purple', marker='s')
axes[1, 0].set_title("Scatter Plot")
for i in range(len(x)):
    axes[1, 0].annotate(f"({x[i]},{y[i]})", (x[i], y[i]))

# 4th subplot: Pie chart
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'gold', 'lightgreen'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
