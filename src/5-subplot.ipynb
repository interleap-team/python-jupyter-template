### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the subplots
x = np.linspace(0, 10, 100)
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Creating the 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line chart in the first subplot
axs[0, 0].plot(x, y, color='blue', label="y = x^2")
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("y")
axs[0, 0].legend()
axs[0, 0].grid(True)

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(axis='y', linestyle="--", linewidth=0.5)

# Scatter plot in the third subplot
axs[1, 0].scatter(scatter_x, scatter_y, color='orange', label="Data Points")
for i in range(len(scatter_x)):
    axs[1, 0].text(scatter_x[i], scatter_y[i], f"({scatter_x[i]}, {scatter_y[i]})", fontsize=9, ha='right', va='bottom')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("x values")
axs[1, 0].set_ylabel("y values")
axs[1, 0].grid(True)
axs[1, 0].legend()

# Pie chart in the fourth subplot
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'blue', 'green', 'purple'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```
