### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 6)
y = x ** 2
categories, values = ['A', 'B', 'C', 'D'], [5, 7, 3, 9]
labels, sizes = ['Python', 'Java', 'C++' 'JavaScript'], [40, 25, 20, 15]
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(x, y, marker='o'); axes[0, 0].set_title("Line Plot")
axes[0, 1].bar(categories, values); axes[0, 1].set_title("Bar Plot")
axes[1, 0].scatter(x, y); axes[1, 0].set_title("Scatter Plot")
for i in range(len(x)):
    axes[1, 0].annotate(f'({x[i]}, {y[i]})', (x[i], y[i]))
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%'); axes[1, 1].set_title("Pie Chart")
plt.tight_layout()
plt.show()
```
