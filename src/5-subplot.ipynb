### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
#coded by Ramashreenidhi k
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.array([1, 2, 3, 4, 5])
y_line = np.array([2, 3, 5, 7, 11])
y_bar = np.array([5, 1, 3, 4])
labels_bar = ['A', 'B', 'C', 'D']
x_scatter = np.array([1, 2, 3, 4, 5])
y_scatter = np.array([2, 4, 6, 8, 10])
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Chart
axs[0, 0].plot(x, y_line, marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X Axis')
axs[0, 0].set_ylabel('Y Axis')

# Bar Chart
axs[0, 1].bar(labels_bar, y_bar, color='lightcoral')
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_ylabel('Values')

# Scatter Plot
axs[1, 0].scatter(x_scatter, y_scatter, color='lightblue')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X Axis')
axs[1, 0].set_ylabel('Y Axis')

# Pie Chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
