### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Data for line plot
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 5, 8, 10]

# Data for bar plot
categories2 = ['A', 'B', 'C', 'D']
values2 = [5, 7, 3, 9]

# Data for scatter plot
x3 = [1, 2, 3, 4, 5]
y3 = [2, 4, 6, 8, 10]

# Data for pie chart
labels4 = ['Python', 'Java', 'C', 'Javascript']
sizes4 = [40, 25, 20, 15]

# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot line chart in the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line Plot')

# Plot bar chart in the second subplot
axs[0, 1].bar(categories2, values2)
axs[0, 1].set_title('Bar Plot')

# Plot scatter plot in the third subplot
axs[1, 0].scatter(x3, y3)
axs[1, 0].set_title('Scatter Plot')

# Plot pie chart in the fourth subplot
axs[1, 1].pie(sizes4, labels=labels4, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
