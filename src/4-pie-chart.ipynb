### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

def create_pie_chart(labels, sizes):
    try:
        if not labels or not sizes:
            raise ValueError("Labels and sizes must not be empty.")
        if len(labels) != len(sizes):
            raise ValueError("Labels and sizes must have the same length.")
        if not all(isinstance(size, (int, float)) for size in sizes):
            raise TypeError("Sizes must be numbers.")
        if sum(sizes) <= 0:
            raise ValueError("Sizes must sum to a positive number.")

        plt.figure(figsize=(8, 8))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
        plt.title('Programming Languages Popularity')
        plt.show()

    except ValueError as ve:
        print(f"ValueError: {ve}")
    except TypeError as te:
        print(f"TypeError: {te}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

create_pie_chart(labels, sizes)
```
